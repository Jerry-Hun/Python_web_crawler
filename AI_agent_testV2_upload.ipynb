```python
# Build up AI agent for Grab stock prices --> plot charts --> calculate values --> Summary by LLM
import os
import numpy as np
import yfinance as yf
import ast
import matplotlib.pyplot as plt
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain.agents import initialize_agent, Tool

# 1️⃣ OpenRouter API Key
os.environ.pop("http_proxy", None)
os.environ.pop("https_proxy", None)
os.environ["OPENROUTER_API_KEY"] = "sk-or-v1-"

# 2️⃣ LLM
llm = ChatOpenAI(
    model="deepseek/deepseek-r1-0528:free",
    base_url="https://openrouter.ai/api/v1",
    api_key=os.getenv("OPENROUTER_API_KEY"),
    temperature=0.1
)

# ────────────────────────────────────
# 3️⃣ Define functions
# ────────────────────────────────────
def fetch_history(args: str):
    """
    args: 'AAPL,2weeks' 或 'AAPL,14d' 或 'AAPL' (預設 14d)
    回傳: (prices_list, dates_list)
    """
    parts = [p.strip() for p in args.split(",", 1)]
    symbol = parts[0]
    period = parts[1].lower() if len(parts) > 1 else "14d"
    if period in {"2weeks", "2week", "two weeks", "two_week", "14d"}:
        period = "14d"
    allowed = {"1d", "5d", "14d", "1mo", "3mo", "6mo", "1y", "2y", "5y", "10y", "ytd", "max"}
    if period not in allowed:
        period = "14d"
    df = yf.Ticker(symbol).history(period=period)
    prices = df["Close"].tolist()
    dates = df.index.astype(str).tolist()
    return prices, dates

def calc_metrics(prices: list):
    """
    輸入: prices list (or string representation of a list)
    回傳: (ann_vol, sharpe, max_dd)
    """
    # Handle string input by converting to list
    prices = safe_to_list(prices)
    if not prices:
        raise ValueError("Invalid prices input: empty or unparseable list")
    arr = np.array(prices, dtype=float)
    # Calculate annualized volatility
    ann_vol = np.std(arr) * np.sqrt(252)
    # Calculate daily returns for Sharpe ratio
    returns = np.diff(arr) / arr[:-1]
    sharpe = (np.mean(returns) * 252) / (ann_vol or 1e-8)
    # Calculate cumulative returns and max drawdown
    cum = np.cumprod(1 + returns)
    dd = (cum - np.maximum.accumulate(cum)) / np.maximum.accumulate(cum)
    max_dd = dd.min() if len(dd) > 0 else 0
    return ann_vol, sharpe, max_dd

def plot_history(data: tuple[list[float], list[str]]) -> str:
    """
    輸入: tuple of (prices, dates) or string representation of such tuple
    回傳: string confirming plot display
    Plot the price history using matplotlib and display the plot.
    """
    # Handle string input by converting to tuple
    if isinstance(data, str):
        try:
            data = ast.literal_eval(data)
        except Exception as e:
            raise ValueError(f"Failed to parse input as tuple: {e}")
    # Ensure input is a tuple with exactly two elements
    if not isinstance(data, tuple) or len(data) != 2:
        raise ValueError("Input must be a tuple of (prices, dates)")
    prices, dates = data
    # Validate that prices and dates are lists
    if not (isinstance(prices, list) and isinstance(dates, list)):
        raise ValueError("Prices and dates must be lists")
    # Simplify dates by removing timestamp and timezone if present
    dates = [d.split()[0] if ' ' in d else d for d in dates]
    plt.figure(figsize=(8, 4))
    plt.plot(dates, prices, marker='o')
    plt.title("AAPL Price History")
    plt.xlabel("Date")
    plt.ylabel("Price (USD)")
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    return "Price history plot displayed successfully."

def explain_func(args: tuple):
    """
    輸入: (symbol, ann_vol) or string representation of such tuple
    回傳: 一句話解釋
    """
    # Handle string input by converting to tuple
    if isinstance(args, str):
        try:
            args = ast.literal_eval(args)
        except Exception as e:
            raise ValueError(f"Failed to parse input as tuple: {e}")
    # Ensure input is a tuple with exactly two elements
    if not isinstance(args, tuple) or len(args) != 2:
        raise ValueError("Input must be a tuple of (symbol, ann_vol)")
    symbol, ann_vol = args
    prompt = ChatPromptTemplate.from_messages([
        ("system", "你是一位專業投資分析師，請用繁體中文回覆。"),
        ("user", f"已抓取 {symbol} 過去兩週的走勢，年化波動率 {ann_vol:.2%}。請一句話說明其主要特徵與投資含意。")
    ])
    resp = llm.invoke(prompt.format(symbol=symbol, ann_vol=ann_vol))
    return resp.content.strip()

def safe_to_list(x):
    """
    Convert string representation of a list/tuple to a list, or return unchanged if already a list.
    """
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except Exception:
            return []
    return x

# ────────────────────────────────────
# 4️⃣ LangChain Tools
# ────────────────────────────────────
tools = [
    Tool(name="get_history", func=fetch_history, description="輸入 'SYMBOL,2weeks' → 回傳 (prices, dates)"),
    Tool(name="calc_metrics", func=calc_metrics, description="輸入 prices list → 回傳 (ann_vol, sharpe, max_dd)"),
    Tool(name="plot_history", func=plot_history, description="輸入 (prices, dates) → 回傳 string confirming plot display"),
    Tool(name="explain", func=explain_func, description="輸入 (symbol, ann_vol) → 回傳一句話解釋")
]

# ────────────────────────────────────
# 5️⃣ AgentExecutor
# ────────────────────────────────────
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent="zero-shot-react-description",
    verbose=True,
    handle_parsing_errors=True
)

# ────────────────────────────────────
# 6️⃣ Agent：Grab 2 weeks stock price → calculate→ Plot charts → LLM generate explains
# ────────────────────────────────────
result = agent.run(
    "請抓取 AAPL 過去兩週的收盤價、計算年化波動率、夏普比、最大回撤，"
    "畫出走勢圖，並一句話解釋其投資含意。"
)
print(result)
```

```python
# Build LLM models with RAG for file Q&A
import os
import numpy as np
import yfinance as yf
from langchain.document_loaders import PyPDFLoader, TextLoader, WebBaseLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.chains import RetrievalQA
from langchain_openai import ChatOpenAI
from langchain.agents import Tool, create_react_agent, AgentExecutor
from langchain.prompts import PromptTemplate
from retry import retry
import logging

# Set up logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# ── Environment ──
os.environ["HF_HUB_DISABLE_SYMLINKS_WARNING"] = "1"  # Suppress HuggingFace symlink warning
os.environ.pop("http_proxy", None)
os.environ.pop("https_proxy", None)
os.environ["OPENROUTER_API_KEY"] = "sk-or-v1-"
os.environ["USER_AGENT"] = "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"

# ── Read file and Chunk ──
logger.info("Loading documents...")
try:
    # Load PDF (AAPL 10-K)
    pdf_loader = PyPDFLoader(r"C:\Users\Jerry Hunter\Downloads\AAPL_10K.pdf")
    pdf_docs = pdf_loader.load()

    # Load Markdown (research.md)
    md_loader = TextLoader(r"C:\Users\Jerry Hunter\Downloads\research.md", encoding="utf8")
    md_docs = md_loader.load()

    # Load mock news data
    news_loader = WebBaseLoader("https://finance.yahoo.com/quote/AAPL/news/")
    news_docs = news_loader.load()

    # Combine all documents
    docs = pdf_docs + md_docs + news_docs

    # Split documents into chunks
    splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
    chunks = splitter.split_documents(docs)
except Exception as e:
    logger.error(f"Error loading documents: {e}")
    raise

# ── Initial LLM with Retry Logic and Fallback ──
@retry(tries=3, delay=2, backoff=2, logger=logger)
def initialize_llm(model="mistralai/mixtral-8x7b-instruct"):
    logger.info(f"Attempting to initialize LLM with model: {model}")
    return ChatOpenAI(
        model=model,
        base_url="https://openrouter.ai/api/v1",
        api_key=os.getenv("OPENROUTER_API_KEY"),
        temperature=0.1,
        default_headers={"User-Agent": os.getenv("USER_AGENT")},
        max_retries=3
    )

try:
    llm = initialize_llm()  # Try primary model
except Exception as e:
    logger.warning(f"Primary model failed: {e}. Falling back to deepseek/deepseek-r1-0528:free")
    try:
        llm = initialize_llm(model="deepseek/deepseek-r1-0528:free")  # Fallback model
    except Exception as e:
        logger.error(f"Failed to initialize LLM after fallback: {e}")
        raise

# ── Embeddings & build Vector Store ──
logger.info("Initializing embeddings and vector store...")
try:
    embeddings = HuggingFaceEmbeddings(
        model_name="BAAI/bge-small-en",
        model_kwargs={"device": "cpu"},
        encode_kwargs={"normalize_embeddings": True}
    )
    vectorstore = FAISS.from_documents(chunks, embeddings)
except Exception as e:
    logger.error(f"Error initializing embeddings/vector store: {e}")
    raise

# ── Define functions as tools ──
def ask_report(query: str) -> str:
    """
    Ask a question about the AAPL 10-K, research.md, or news data.
    Input: query string
    Returns: Answer with source snippets
    """
    logger.info(f"Running ask_report with query: {query}")
    try:
        qa_chain = RetrievalQA.from_chain_type(
            llm=llm,
            chain_type="stuff",
            retriever=vectorstore.as_retriever(search_kwargs={"k": 3}),
            return_source_documents=True
        )
        result = qa_chain.invoke({"query": query})
        answer = result["result"]
        sources = []
        for doc in result["source_documents"]:
            snippet = doc.page_content[:200].replace('\n', ' ')
            sources.append(f"- {doc.metadata['source']}: {snippet}…")
        sources_text = "\n".join(sources)
        return f"Answer: {answer}\n\nSources:\n{sources_text}"
    except Exception as e:
        logger.error(f"Error in ask_report: {e}")
        return f"Error: Failed to process query: {e}"

def calc_vol(args: str) -> float:
    """
    Calculate annualized volatility for a stock.
    Input: 'SYMBOL,period' (e.g., 'AAPL,2weeks')
    Returns: Annualized volatility
    """
    logger.info(f"Running calc_vol with args: {args}")
    try:
        parts = [p.strip() for p in args.split(",", 1)]
        symbol = parts[0]
        period = parts[1].lower() if len(parts) > 1 else "14d"
        if period in {"2weeks", "2week", "two weeks", "two_week", "14d"}:
            period = "14d"
        allowed = {"1d", "5d", "14d", "1mo", "3mo", "6mo", "1y", "2y", "5y", "10y", "ytd", "max"}
        if period not in allowed:
            period = "14d"
        df = yf.Ticker(symbol).history(period=period)
        prices = df["Close"].tolist()
        arr = np.array(prices, dtype=float)
        ann_vol = np.std(arr) * np.sqrt(252)
        return ann_vol
    except Exception as e:
        logger.error(f"Error in calc_vol: {e}")
        raise

# ── LangChain Tools ──
tools = [
    Tool(
        name="ask_report",
        func=ask_report,
        description="輸入問題以查詢 AAPL 10-K、研究報告或新聞資料，輸出回答與來源片段"
    ),
    Tool(
        name="calc_vol",
        func=calc_vol,
        description="輸入 'SYMBOL,period' (如 'AAPL,2weeks')，回傳年化波動率"
    )
]

# ── AgentExecutor ──
logger.info("Initializing agent...")
try:
    # Define the prompt template for the ReAct agent
    react_prompt = PromptTemplate.from_template("""
    Answer the following questions as best you can. You have access to the following tools:

    {tools}

    Use the following format:

    Question: the input question you must answer
    Thought: you should always think about what to do
    Action: the action you should take, should be one of [{tool_names}]
    Action Input: the input to the action
    Observation: the result of the action
    ... (this Thought/Action/Action Input/Observation can repeat N times)
    Thought: I now know the final answer
    Final Answer: the final answer to the original input question

    Begin!

    Question: {input}
    Thought: {agent_scratchpad}
    """)

    agent = create_react_agent(llm=llm, tools=tools, prompt=react_prompt)
    agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True, handle_parsing_errors=True, max_iterations=10)
except Exception as e:
    logger.error(f"Error initializing agent: {e}")
    raise

# ── Run Q&A and finals from report ──
query = "請問 AAPL 10-K 報告中列出的主要風險因素，並計算 AAPL 過去兩週的年化波動率。"
logger.info(f"Running agent with query: {query}")
try:
    result = agent_executor.invoke({"input": query})
    print("── 最終結果 ──")
    print(result["output"])
except Exception as e:
    logger.error(f"Error running agent: {e}")
    raise
```

```python
# Build Gradio interactive with LLM and Charts
import os
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
import gradio as gr
from langchain_openai import ChatOpenAI
from langchain import PromptTemplate, LLMChain
import io
from PIL import Image

# Set environment variables
os.environ.pop("http_proxy", None)
os.environ.pop("https_proxy", None)
os.environ["OPENAI_API_KEY"] = os.getenv("OPENROUTER_API_KEY")

# Initialize LLM
llm = ChatOpenAI(
    model="deepseek/deepseek-r1-0528:free",
    base_url="https://openrouter.ai/api/v1",
    openai_api_key=os.getenv("OPENAI_API_KEY"),
    temperature=0.2,
    max_retries=2,
)

# Prepare summary prompt and chain
summary_prompt = PromptTemplate(
    input_variables=["ticker", "period"],
    template=(
        "請根據 {ticker} 在過去 {period} 的走勢，"
        "用繁體中文寫一段不超過 300 字的投資主題摘要：\n"
        "- 策略概覽\n"
        "- 風險要點\n"
        "- 建議操作"
    )
)
summary_chain = LLMChain(llm=llm, prompt=summary_prompt)

# Core function: Fetch data, generate summary, plot chart, and answer questions
def generate_report(ticker: str, period: str, question: str):
    # 1) Generate investment summary
    try:
        res = summary_chain.invoke({"ticker": ticker, "period": period})
        summary = res["text"]  # Access 'text' key
    except Exception as e:
        summary = f"Error generating summary: {e}"

    # 2) Fetch data and generate chart as image
    try:
        df = yf.Ticker(ticker).history(period=period)
        if df.empty:
            raise ValueError(f"No data returned for {ticker}")
        prices = df["Close"].tolist()
        dates = df.index.astype(str).tolist()

        # Set font for Chinese characters (Windows-compatible)
        plt.rcParams['font.sans-serif'] = ['Microsoft YaHei', 'SimHei', 'Arial', 'sans-serif']
        plt.rcParams['axes.unicode_minus'] = False  # Handle minus sign

        fig, ax = plt.subplots(figsize=(8, 4))
        ax.plot(dates, prices, marker='o')
        ax.set_title(f"{ticker} 收盤價 ({period})")
        ax.set_xlabel("日期")
        ax.set_ylabel("價格 (USD)")
        ax.grid(True)
        plt.xticks(rotation=45)
        plt.tight_layout()

        # Save plot to bytes buffer instead of showing it
        buf = io.BytesIO()
        fig.savefig(buf, format="png")
        buf.seek(0)
        img = Image.open(buf)
        plt.close(fig)  # Close the figure to free memory
    except Exception as e:
        summary += f"\nError fetching {ticker} data: {e}"
        img = None

    # 3) Handle advanced question
    answer = "（沒有進階提問）"
    if question and question.strip():
        qa_prompt = PromptTemplate(
            input_variables=["summary", "question", "ticker", "period"],
            template=(
                "以下是 {ticker} 過去 {period} 的走勢簡述：\n"
                "{summary}\n\n"
                "請回答：{question}"
            )
        )
        qa_chain = LLMChain(llm=llm, prompt=qa_prompt)
        try:
            qa_res = qa_chain.invoke({
                "summary": summary,
                "question": question,
                "ticker": ticker,
                "period": period
            })
            answer = qa_res["text"]
        except Exception as e:
            answer = f"Error answering question: {e}"

    return summary, img, answer

# Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("## 🎯 AI 投資報告助理 V1 (Gradio)")
    with gr.Row():
        ticker_input = gr.Textbox(label="股票代號", value="AAPL")
        period_input = gr.Dropdown(
            label="期間",
            choices=["5d", "14d", "1mo", "3mo"],
            value="1mo"
        )
    question_input = gr.Textbox(label="進階提問 (可留空)")
    summary_out = gr.Textbox(label="📄 投資主題摘要")
    plot_out = gr.Image(label="📈 價格走勢圖")
    answer_out = gr.Textbox(label="❓ 進階問答")
    submit_btn = gr.Button("生成報告")

    submit_btn.click(
        fn=generate_report,
        inputs=[ticker_input, period_input, question_input],
        outputs=[summary_out, plot_out, answer_out]
    )

if __name__ == "__main__":
    demo.launch()
```
